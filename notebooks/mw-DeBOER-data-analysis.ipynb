```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('DATA/De Boer inventaris_backup 20180223.xlsx')
```

```python
df['date'] = df.date.replace({'00-': ''}, regex=True)
df['date'] = df.date.replace({'1066-': '1966'}, regex=True)
```

```python
df['1958':'1969'].resample('12M').count()
```

```python
df['date'] = pd.to_datetime(df['date'], errors = 'coerce')
```

```python
df = df.set_index(pd.DatetimeIndex(df['date']))
```

```python
df['place'].resample('12M').count().plot()
plt.savefig('yearly.png', dpi=300)
```

```python
places = df.place.value_counts()
```

```python
x = places.index[:20]
y = places.values[:20]
```

```python
bar_chart = sns.barplot(x, y)
loc, labels = plt.xticks()
bar_chart.set_xticklabels(labels, rotation=90)
plt.show;
```
