```python
import pandas as pd
import seaborn as sns
import os
import matplotlib.pyplot as plt
```

```python
data_dir = '../data/processed/train/'
figures_dir = '../output/figures/'
```

```python
# Empty lists
categories = []
img_categories = []
n_train = []
n_valid = []
#n_test = []


# Iterate through each category
for d in os.listdir(data_dir):
    if os.path.isdir(data_dir + d):

        if len(os.listdir(data_dir + d)) > 0:
        
            categories.append(d)

            # Number of each image
            train_imgs = os.listdir(data_dir + d)
            n_train.append(len(train_imgs))
            n_train.append(len(train_imgs))




# Dataframe of categories
cat_df = pd.DataFrame({'category': categories,
                       'n_train': n_train,
                       }).sort_values('category')


cat_df.sort_values('n_train', ascending=False, inplace=True)
```

```python
ax = cat_df[cat_df['n_train'] > 9].plot(x='category', y='n_train', kind='bar', figsize=(20,5))
plt.tight_layout()
ax.get_legend().remove()
plt.savefig(figures_dir + 'categories.png', dpi=300)
plt.show()
```
